```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
training_data=pd.read_csv("Gotem Pumpkins.csv")
training_data=training_data.drop("Unnamed: 0",axis=1)
training_data["Class"]=training_data["Class"].map({'Ürgüp Sivrisi': 0, 'Çerçevelik': 1})
training_data.head()
```

```python
testing_data=pd.read_csv("Freyja_Pumpkins.csv")
testing_data=testing_data.drop("Unnamed: 0",axis=1)
testing_data["Class"]=testing_data["Class"].map({'Ürgüp Sivrisi': 0, 'Çerçevelik': 1})
testing_data.head()
```

```python
training_scaled_features = training_data.columns.drop('Class') 
training_data[training_scaled_features] = (training_data[training_scaled_features] - training_data[training_scaled_features].mean()) / training_data[training_scaled_features].std()
training_data.head()
```

```python
testing_scaled_features = testing_data.columns.drop('Class') 
testing_data[testing_scaled_features] = (testing_data[testing_scaled_features] - testing_data[testing_scaled_features].mean()) / testing_data[testing_scaled_features].std()
testing_data.head()
```

```python
X_train=training_data.drop('Class',axis=1)
Y_train=training_data['Class']
X_train=X_train.to_numpy()
Y_train=Y_train.to_numpy()
```

```python
X_test=testing_data.drop('Class',axis=1)
Y_test=testing_data['Class']
X_test=X_test.to_numpy()
Y_test=Y_test.to_numpy()
```

```python

# Plot each feature against the CLass 
plt.figure(figsize=(15, 20))

for i,col in enumerate(training_data.columns.drop('Class'),1):
    plt.subplot(4, 3, i)
    plt.scatter(training_data[col], training_data['Class'], alpha=0.6, edgecolor='k')
    plt.title(col, fontsize=12)
    plt.xlabel(col, fontsize=10)
    plt.ylabel("Class", fontsize=10)
    plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def logistic_regression(x, y, learning_rate, iterations):
    row,f=x.shape
    weights=np.zeros(f)
    bias=0
    cost_list=[]
    for i in range(iterations):
        z=x.dot(weights)+bias
        y_pred=sigmoid(z)
        cost=-y*np.log(y_pred)+(y-1)*np.log(1-y_pred)
        
        total_cost=cost.sum()/row
        cost_list.append(total_cost)
        dw=(x.T.dot((y_pred-y)))/row
        db=(np.sum(y_pred-y))/row
        weights-=learning_rate*dw
        bias -= learning_rate * db

    return weights, bias, cost_list
        
```

```python
Iterations=10000
learning_rate=0.1
```

```python
weights,bias,cost_list=logistic_regression(X_train, Y_train,learning_rate,Iterations)
```

```python
print(weights,bias)
```

```python
def predict(X, weights, bias):
    z =X.dot(weights) + bias
    y_pred = sigmoid(z) 
    return [1 if i > 0.5 else 0 for i in y_pred]
    
```

```python
y_pred_train = predict(X_train, weights, bias) 
y_pred_test = predict(X_test, weights, bias) 

train_accuracy = np.mean(y_pred_train == Y_train) * 100
test_accuracy = np.mean(y_pred_test == Y_test) * 100
print(f"Train Accuracy: {train_accuracy:.2f}%")
print(f"Test Accuracy: {test_accuracy:.2f}%")
```
